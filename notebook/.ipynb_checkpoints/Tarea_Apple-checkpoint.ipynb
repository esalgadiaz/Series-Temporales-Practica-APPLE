# HW 01: Predicción Ingresos Apple

### Importación de librerías

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm
from sktime.forecasting.base import ForecastingHorizon
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.ets import AutoETS
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.transformations.series.boxcox import LogTransformer
from sktime.utils.plotting import plot_correlations
from sktime.forecasting.arima import AutoARIMA
import warnings
```

### Configuracion de las figuras y advertencias:

```python
sns.set(rc={'figure.figsize':(11, 4)})
warnings.filterwarnings('ignore')
```

### Carga de datos

Selección de los ingresos de Apple desde el T1 de 1990 hasta el T2 de 2021 con las fechas como índice:

```python
apple = (pd.read_csv('aapl_ok.csv', sep=';', usecols=['fecha','Ingresos'], skiprows=[1])
    .assign(**{'fecha': lambda df: pd.to_datetime(df['fecha'],format="%m/%d/%Y")})
    .assign(**{'Ingresos': lambda df: df['Ingresos'].str.replace(',', '.').astype(float)})
)
apple_df = apple.set_index('fecha')
apple_df = apple_df.resample("q").last() # Aunque ya sean datos trimestrales, lo configuro para modificar el tipo de dato.
apple_df.head()
```

```python
apple_df['Ingresos']
```

#### Representación preliminar

```python
plot_series(apple_df['Ingresos'].astype('float64').to_period('Q'), labels=["Ingresos"])
```

Pregunta 1: ¿Los ingresos de Apple son estacionarios? ¿En media? ¿En varianza?
 
Respuesta 1: Los ingresos de Apple no son estacionarios ni en media ni en varianza debido a que no permanecen constantes a lo largo de la serie.

Pregunta 2:  ¿Cuál es su tendencia? ¿y su componente estacional?

Respuesta 2: A priori, parece que las componentes de tendencia y estacionalidad son multiplicativas (tendencia en forma exponencial -aunque no muy pronunciada-, y oscilación creciente); no obstante, se comprobará qué modelos ets y arima se ajustan mejor.

### Análisis de la estacionalidad

* Hay componente estacional
* Más ventas en el T4
* El Componente estacional no es estacionario

```python
ax = plt.gca()
sm.graphics.tsa.quarter_plot(apple_df['Ingresos'],ax=ax)
ax.set_title('Comportamiento Estacional')
```

## Modelo ETS

#### Separación del dataset: Train-Test

```python
y_train, y_test = temporal_train_test_split(y = apple_df['Ingresos'].astype('float64').to_period('Q'), test_size=8)
# plotting for illustration
plot_series(y_train, y_test, labels=["y_train", "y_test"])
print(y_train.shape[0], y_test.shape[0])
```

#### Longitud del pronóstico o predicción

```python
fh = np.arange(1, 9)
```

### Modelo ETS Automático

```python
apple_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
```

#### Entrenamiento del modelo

```python
apple_auto_model.fit(y_train)
```

#### Características del modelo:

```python
print(apple_auto_model.summary())
```

De manera automática, el modelo considera que la tendencia es aditiva y el componente estacional multiplicativo.

### Predicción del modelo ETS (MAM)

```python
apple_pred_ets = apple_auto_model.predict(fh)
```

#### Intervalos de confianza

```python
apple_pred_ints = apple_auto_model.predict_interval(fh, coverage=0.90)
apple_pred_ints
```

#### Representación de la predicción vs realidad

```python
plot_series(y_train, apple_pred_ets, y_test, labels=["Apple", "Apple pred", "Apple REAL"])
```

#### Representación de la predicción vs realidad + intervalos

```python
fig, ax = plot_series(y_train, apple_pred_ets, y_test, labels=["Apple", "Apple pred", "Apple REAL"])
ax.fill_between(
    ax.get_lines()[1].get_xdata(),
    apple_pred_ints[('Coverage', 0.9, 'lower')],
    apple_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[1].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left')
```

### Precisión del modelo ETS (MAM)

Error absoluto medio porcentual

```python
mape_ets = mean_absolute_percentage_error(y_test, apple_pred_ets)
mape_ets
```

Error cuadrático medio

```python
mse_ets = MeanSquaredError()
mse_ets = mse_ets(y_test, apple_pred_ets)
mse_ets
```

Raíz del error cuadrático medio

```python
rmse_ets = MeanSquaredError(square_root=True)
rmse_ets = rmse_ets(y_test, apple_pred_ets)
rmse_ets
```

### Modelo ETS $\to$ Tendencia multiplicativa y error aditivo

```python
apple_ets_model = AutoETS(auto=False, trend='mul', damped_trend=False,
                          seasonal='mul', error='add', sp=4, n_jobs=-1)
```

#### Entrenamiento del modelo

```python
apple_ets_model.fit(y_train)
```

#### Características del modelo

```python
print(apple_ets_model._fitted_forecaster.summary())
```

### Predicción del modelo ETS (AMM)

```python
apple_pred_ets2 = apple_ets_model.predict(fh)
```

#### Representación de la predicción vs realidad

```python
plot_series(y_train, apple_pred_ets2, y_test, labels=["Apple", "Apple pred", "Apple REAL"])
```

### Precisión del modelo ETS (AMM)

Error absoluto medio porcentual

```python
mape_ets2 = mean_absolute_percentage_error(y_test, apple_pred_ets2)
mape_ets2
```

Error cuadrático medio

```python
mse_ets2 = MeanSquaredError()
mse_ets2 = mse_ets2(y_test, apple_pred_ets2)
mse_ets2
```

Raíz del error cuadrático medio

```python
rmse_ets2 = MeanSquaredError(square_root=True)
rmse_ets2 = rmse_ets2(y_test, apple_pred_ets2)
rmse_ets2
```

Respuesta 2: Se puede comprobar que considerando una tendencia multiplicativa se consigue un mejor ajuste del modelo, mejorando así la predicción.

## Modelo ARIMA

#### Tranformación logarítmica

Como la serie no presenta estacionariedad en varianza, se aplica una tranformación logarítmica a los ingresos:

```python
transformer = LogTransformer()
log_apple_ts= transformer.fit_transform(apple_df['Ingresos'].astype('float64').to_period('Q'))
log_apple_ts.tail()
```

#### Representación preliminar: original vs log

```python
fig, ax =plot_series(apple_df['Ingresos'].astype('float64').to_period('Q'), labels=["Ingresos - Original"])
ax.set_title('Ingresos Apple')

fig, ax =plot_series(log_apple_ts, labels=["Ingresos - Log"])
ax.set_title('Ingresos Apple: Transformación LOG')
```

#### Autocorrelación de la serie (Logs)

```python
plot_correlations(log_apple_ts)
```

#### Separación del dataset: Train-Test

```python
y_train, y_test = temporal_train_test_split(y =apple_df['Ingresos'].astype('float64').to_period('Q'), test_size=8)
log_y_train, log_y_test = temporal_train_test_split(y = log_apple_ts, test_size=8)
```

#### Longitud del pronóstico o predicción

```python
fh = np.arange(1, 9)
```

### Modelo ARIMA Auto

```python
forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_y_train)
```

#### Características del modelo

```python
print(forecaster.summary())
```

### Predicción del modelo ARIMA

```python
log_y_pred = forecaster.predict(fh)
```

#### Reversión logarítmica

```python
apple_pred_arima = np.exp(log_y_pred)
```

#### Representación de la predicción vs realidad

```python
plot_series(y_train, apple_pred_arima, y_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
```
