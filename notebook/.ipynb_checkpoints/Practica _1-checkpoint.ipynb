```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from matplotlib import pyplot as plt
from sktime.utils.plotting import plot_series
from datetime import datetime
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.ets import AutoETS
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.arima import AutoARIMA
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.utils.plotting import plot_correlations
from sktime.transformations.series.difference import Differencer
import warnings
warnings.filterwarnings('ignore')
```

```python
df_apple = (pd.read_csv('C:/Users/Pablo/Desktop/Entornos/prediccion/data/raw/aapl_ok.csv',sep=';') 
           .assign(**{'fecha': lambda df: pd.to_datetime(df['fecha'])})
           .assign(**{'Ingresos': lambda df: df['Ingresos'].str.replace(',', '.').astype(float)}))
```

```python
pd.isnull(df_apple[[x for x in df_apple.columns[pd.isna(df_apple).any()].tolist()]]).sum()
```

```python
#modificamos la columna trimestre para que muestre el trimestre real y no el fiscal de la compañía 
#y establecemos el trimestre como índice

df_apple['Trimeste'] = pd.PeriodIndex(df_apple.fecha, freq='Q')
df_apple = df_apple.set_index('Trimeste')

#creamos un nuevo data frame con los trimestres como indices e ingresos como datos, por tanto eliminamos la fehca
df_apple= df_apple.drop('fecha', axis=1)
df_apple.head()
```

```python
# df_apple_train = df_apple.iloc[0:119]
# df_apple_train = df_apple_train.set_index('Trimeste')
# df_apple_train.tail()
```

```python
fig, ax = plot_series(df_apple, labels=["Ingresos trimestrales"])
ax.set_title('Ingresos Apple trimestrales')
```

```python
df_apple_train, df_apple_test = temporal_train_test_split(y = df_apple['Ingresos'].astype('float64'), test_size=8)
# we will try to forecast y_test from y_train
# plotting for illustration
plot_series(df_apple_train, df_apple_test, labels=["df_apple_train", "df_apple_test"])
print(df_apple_train.shape[0], df_apple_test.shape[0])
```

# Modelo ETS

```python
# step 2: specifying forecasting horizon
fh = np.arange(1, 17)

# step 3: specifying the forecasting algorithm
# sp 4 por los Q
apple_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
```

```python
# para los datos de train
apple_auto_model.fit(df_apple_train)
```

```python
# es un modelo MAM error multiplicativo tendecia lineal aditiva
print(apple_auto_model.summary())
```

```python

# step 5: querying predictions
apple_pred = apple_auto_model.predict(fh)
print(apple_pred)
```

```python
apple_pred_ints = apple_auto_model.predict_interval(fh, coverage=0.9)
apple_pred_ints
```

```python
# optional: plotting predictions and past data
plot_series(df_apple_train, apple_pred,df_apple_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
```

```python
fig, ax = plot_series(df_apple_train, apple_pred, df_apple_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
ax.fill_between(
    ax.get_lines()[-2].get_xdata(),
    apple_pred_ints[('Coverage', 0.9, 'lower')],
    apple_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-2].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left')
```

```python
plot_series(df_apple_train[100:], apple_pred,df_apple_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
```

# MODELO ARIMA

```python
df_apple = df_apple['Ingresos'].astype('float64').sort_index()
df_apple.tail()
```

```python
# Apply Log Transformer

transformer = LogTransformer()
log_apple= transformer.fit_transform(df_apple)
log_apple.tail()
```

```python
# Plot Log Data
ax = log_apple.plot(marker='o', linestyle='-')
ax.set_ylabel('Ingresos de APPLE')
ax.set_title('Ingresos de APPLE: Transformación LOG')
```

```python
# Plot Log Data
fig, ax =plot_series(log_apple, labels=["Ventas"])
ax.set_title('Ingresos de APPLE: Transformación LOG')
```

```python
# Plot Log Data & Original Data
fig, ax =plot_series(df_apple, labels=["Ventas"])
ax.set_title('Ventas Coca Cola: Serie Original')
fig, ax =plot_series(log_apple, labels=["Ventas"])
ax.set_title('Ventas Coca Cola: Transformación LOG')
```

```python
# Autocorrelation Fuction Original Time Series
plot_correlations(log_apple)
```

```python
# Autocorrelation Fuction
# d=1; D=0; S=4
# 
transf_diff=Differencer(lags=[1])
plot_correlations(transf_diff.fit_transform(log_apple))
```

```python
# Autocorrelation Fuction
# d=0; D=1; S=4
# 
transf_diff=Differencer(lags=[4])
plot_correlations(transf_diff.fit_transform(log_apple))
```

```python
# Autocorrelation Fuction
# d=1; D=1; S=4
# 
transf_diff=Differencer(lags=[1,4])
plot_correlations(transf_diff.fit_transform(log_apple))
```

## Modelos ARIMA

```python
# Select Train & test sample
# we will try to forecast y_test from y_train
# plotting for illustration
# original and log samples
df_apple_train , df_apple_test = temporal_train_test_split(y =df_apple, test_size=8)
log_apple_train, log_apple_test = temporal_train_test_split(y =log_apple, test_size=8)
plot_series(df_apple_train , df_apple_test, labels=["Train", "Test"])
# Time Series Size
print(df_apple_train.shape[0], df_apple_test.shape[0])
```

```python
# Forecast Horizon
fh = np.arange(len(df_apple_test)) + 1  # forecasting horizon
fh
```

```python
#  Auto ARIMA Model

forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_apple_train)
```

```python
#  Auto ARIMA Model Summary
print(forecaster.summary())
```

```python
#  Auto ARIMA Forecast
log_apple_pred = forecaster.predict(fh)
log_apple_pred
```

```python
#  Auto ARIMA Forecast
# Original Time series (Invert log transformation)
np.exp(log_apple_pred)
```

```python
# Forecast Accuracy
# MAPE
mean_absolute_percentage_error(log_apple_test, log_apple_pred)
```

```python
# Forecast Accuracy
# MAPE 
# Orignal Time Serie

mean_absolute_percentage_error(df_apple_test, np.exp(log_apple_pred))
```

```python
# Forecast Accuracy
#Mean Squared Error
rmse = MeanSquaredError(square_root=True)
rmse(log_apple_test, log_apple_pred)
```

```python
# Forecast Accuracy
#Mean Squared Error
rmse = MeanSquaredError(square_root=True)
rmse(df_apple_test, np.exp(log_apple_pred))
```

```python
# Forecast Accuracy Plot
# Plotting predictions and past data
plot_series(df_apple_train , np.exp(log_apple_pred),df_apple_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
```

```python
# Forecast Accuracy Plot
# Plotting predictions and past data
# Zoom 2013 -2021
plot_series(df_apple_train[50:], np.exp(log_apple_pred),df_apple_test, labels=["APPLE", "APPLE pred", "APPLE REAL"])
```
