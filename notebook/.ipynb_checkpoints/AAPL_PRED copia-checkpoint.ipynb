## Cargamos las librerías 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from matplotlib import pyplot as plt
from sktime.utils.plotting import plot_series
from datetime import datetime
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.ets import AutoETS
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.arima import AutoARIMA
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.utils.plotting import plot_correlations
from sktime.transformations.series.difference import Differencer
import warnings
warnings.filterwarnings('ignore')
```

## Carga de los datos, tratamiento y limpieza

```python
df_AAPL = (pd.read_csv("/Users/eduardo/Documents/CUNEF_ASIGNATURAS/SERIES_TEMPORALES/Trabajo_HW/data/raw/aapl_ok.csv",sep=';') 
           .assign(**{'fecha': lambda df: pd.to_datetime(df['fecha'])})
           .assign(**{'Ingresos': lambda df: df['Ingresos'].str.replace(',', '.').astype(float)}))
```

```python
pd.isnull(df_AAPL[[x for x in df_AAPL.columns[pd.isna(df_AAPL).any()].tolist()]]).sum()
```

```python
df_AAPL['Trimeste'] = pd.PeriodIndex(df_AAPL.fecha, freq='Q')
df_AAPL = df_AAPL.set_index('Trimeste')

#creamos un nuevo data frame con los trimestres como indices e ingresos como datos, por tanto eliminamos la fehca
df_AAPL= df_AAPL.drop('fecha', axis=1)
df_AAPL.head()
```

Representamos la gráfica de los ingresos por cuatrimestre de la empresa APPLE.

```python
transformer = LogTransformer()
log_AAPL= transformer.fit_transform(df_AAPL)
log_AAPL.tail()
```

```python
# Plot Log Data & Original Data
fig, ax =plot_series(df_AAPL, labels=["Ingresos"])
ax.set_title('Ventas AAPL: Serie Original')
fig, ax =plot_series(log_AAPL, labels=["Ingresos"])
ax.set_title('Ventas AAPL: Transformación LOG')
```

Comportamiento estacional de los cuatrimestres en versión cuatrimestres.

```python
ax = plt.gca()
sm.graphics.tsa.quarter_plot(df_AAPL['Ingresos'],ax=ax)
ax.set_title('Comportamiento Estacional')
```

## Dividimos entre train y test 

Como nos comenta el problema a resolver separamos entre train y test desde los ultimos ocho trimestres, es decir, los dos ultimos años del dataset

```python
df_AAPL_train, df_AAPL_test = temporal_train_test_split(y = df_AAPL['Ingresos'].astype('float64'), test_size=8)
plot_series(df_AAPL_train, df_AAPL_test, labels=["df_AAPL_train", "df_AAPL_test"])
print(df_AAPL_train.shape[0], df_AAPL_test.shape[0])
```

# Modelo ETS

```python
# step 2: specifying forecasting horizon
fh = np.arange(1, 17)

# step 3: specifying the forecasting algorithm
# sp 4 por los Q
AAPL_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
```

```python
# para los datos de train
AAPL_auto_model.fit(df_AAPL_train)
```

```python
# es un modelo MAM error multiplicativo tendecia lineal aditiva
print(AAPL_auto_model.summary())
```

```python
# step 5: querying predictions
AAPL_pred = AAPL_auto_model.predict(fh)
print(AAPL_pred)
```

```python
AAPL_pred_ints = AAPL_auto_model.predict_interval(fh, coverage=0.9)
AAPL_pred_ints
```

```python
# optional: plotting predictions and past data
plot_series(df_AAPL_train, AAPL_pred,df_AAPL_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```

```python
fig, ax = plot_series(df_AAPL_train, AAPL_pred, df_AAPL_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
ax.fill_between(
    ax.get_lines()[-2].get_xdata(),
    AAPL_pred_ints[('Coverage', 0.9, 'lower')],
    AAPL_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-2].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left')
```

```python
plot_series(df_AAPL_train[100:], AAPL_pred,df_AAPL_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```

```python
fig, ax = plot_series(df_AAPL_train[100:], AAPL_pred, df_AAPL_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
ax.fill_between(
    ax.get_lines()[-2].get_xdata(),
    AAPL_pred_ints[('Coverage', 0.9, 'lower')],
    AAPL_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-2].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left');
```

```python
mean_absolute_percentage_error(df_AAPL_test, AAPL_pred[0:8])
```

```python
mse = MeanSquaredError()
mse(df_AAPL_test, AAPL_pred[0:8])
```

```python
rmse = MeanSquaredError(square_root=True)
rmse(df_AAPL_test, AAPL_pred[0:8])
```

```python
# step 2: specifying forecasting horizon
fh = np.arange(1, 7)

# step 3: specifying the forecasting algorithm
AAPL_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)

y = df_AAPL['Ingresos'].astype('float64')

AAPL_auto_model.fit(y)

print(AAPL_auto_model.summary())
```

```python
# step 5: querying predictions
AAPL_pred = AAPL_auto_model.predict(fh)
print(AAPL_pred)
```

```python
plot_series(y, AAPL_pred, labels=["AAPL", "AAPL pred"])
```

```python
# solo componente estacional sin tendencia
forecaster = ExponentialSmoothing(trend=None,seasonal="mul", sp=4)
forecaster.fit(y)
y_pred = forecaster.predict(fh)
plot_series(y, y_pred, labels=["AAPL", "AAPL pred"])
```

```python
#
forecaster = ExponentialSmoothing(trend=None,seasonal="add", sp=4)
forecaster.fit(y)
y_pred = forecaster.predict(fh)
plot_series(y, y_pred, labels=["AAPL", "AAPL pred"])
```

```python

forecaster = ExponentialSmoothing(trend="add",seasonal="add",damped_trend=False, sp=4)
forecaster.fit(y)
y_pred = forecaster.predict(fh)

forecaster = ExponentialSmoothing(trend="add",seasonal="add",damped_trend=True, sp=4)
forecaster.fit(y)
y_pred_dump = forecaster.predict(fh)
plot_series(y[75:], y_pred, y_pred_dump,labels=["AAPL", "AAPL pred","AAPL Pred Dumpeded"])
```

## Predicción modelo ARIMA

```python
# Select Train & test sample
# we will try to forecast y_test from y_train
# plotting for illustration
# original and log samples
df_AAPL_train , df_AAPL_test = temporal_train_test_split(y =df_AAPL, test_size=8)
log_AAPL_train, log_AAPL_test = temporal_train_test_split(y =log_AAPL, test_size=8)
plot_series(df_AAPL_train , df_AAPL_test, labels=["Train", "Test"])
# Time Series Size
print(df_AAPL_train.shape[0], df_AAPL_test.shape[0])
```

```python
#  Auto ARIMA Model

forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_AAPL_train)
```

```python
#  Auto ARIMA Model Summary
print(forecaster.summary())
```

```python
#  Auto ARIMA Forecast
log_AAPL_pred = forecaster.predict(fh)
log_AAPL_pred
```

```python
#  Auto ARIMA Forecast
# Original Time series (Invert log transformation)
np.exp(log_AAPL_pred)
```

```python
# Forecast Accuracy
# MAPE
mean_absolute_percentage_error(log_AAPL_test, log_AAPL_pred)
```

```python
# Forecast Accuracy
#Mean Squared Error
rmse = MeanSquaredError(square_root=True)
rmse(log_AAPL_test, log_AAPL_pred)
```

```python
df_AAPL = (pd.read_csv("/Users/eduardo/Documents/CUNEF_ASIGNATURAS/SERIES_TEMPORALES/Trabajo_HW/data/raw/aapl_ok.csv",sep=';') 
           .assign(**{'fecha': lambda df: pd.to_datetime(df['fecha'])})
           .assign(**{'Ingresos': lambda df: df['Ingresos'].str.replace(',', '.').astype(float)}))
```

```python
separar_fecha = df_AAPL['fecha'].dt.year 
col_ingresos = df_AAPL['Ingresos']
tot_col = pd.concat([separar_fecha,col_ingresos],axis = 1)
DF_TOTAL=tot_col.groupby(['fecha'], sort=True)['Ingresos'].sum()
df_ingXaño = pd.DataFrame(DF_TOTAL)
df_ingXaño.head()
```

Esto son los ingresos totales de la empresa AAPL por año de los siguientes ejercicios, debemos hacer inciso en dos años el 1989 y el año 2021, que son dos valores totales incompletos, es decir, no podremos tener en cuenta los valores que nos dan y debemos no tomar esas dos anualidades como el total de los ingresos.
